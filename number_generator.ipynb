```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
all_data = []

for i in range(4):
    data = np.random.normal(50, 20, 100).astype(int)
    data.sort()
    all_data.append(data)

plt.boxplot(all_data)
plt.xlabel('Data Sets')
plt.ylabel('Value')
plt.title('Data Distribution')
plt.show()
```

```python
for i in range(len(all_data)):
    for j in range(100):
        print(all_data[i][j], end=", ")
    print()
```

```python
# save each data set in its own text file with a number in each line
for i in range(len(all_data)):
    np.savetxt("data/SameMedian/{}.txt".format(i), all_data[i], fmt="%d")
```
